```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gdp = pd.read_csv('gdp.csv') #read gdp csv
```

```python
sns.lmplot(data = gdp, x = 'Ratio', y = 'GDP', hue = 'Region', ci =None) #plot ratio to gdp scatterplot with fits for each Region
plt.title('Economic Prosperity Relative to Trade Status by Region')
plt.xlabel('Exports to Imports Ratio')
plt.ylabel('Real GDP per Capita (PPP) in US Dollars')
plt.figtext(.8, .01, 'Ratio visibly correlated with GDP in Europe and Asia', 
            horizontalalignment='left', wrap = True, fontsize = 8) #caption in bottom right corner
plt.savefig('eda1.pdf', bbox_inches='tight')
plt.show()
```
