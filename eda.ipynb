```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
import kaleido
```

```python
gdp = pd.read_csv('gdp.csv') #read gdp csv
```

```python
sns.lmplot(data = gdp, x = 'Ratio', y = 'GDP', scatter_kws = {'s': 10}, ci =None)
plt.title('Economic Prosperity Relative to Trade Status')
plt.xlabel('Exports to Imports Ratio')
plt.ylabel('Real GDP per Capita (PPP) in US Dollars')
plt.savefig('trade.png', bbox_inches='tight')
plt.show()
```

```python
sns.lmplot(data = gdp, x = 'Ratio', y = 'GDP', hue = 'Region', scatter_kws = {'s': 10}, ci =None)
plt.title('Economic Prosperity Relative to Trade Status by Region')
plt.xlabel('Exports to Imports Ratio')
plt.ylabel('Real GDP per Capita (PPP) in US Dollars')
plt.savefig('traderegion.png', bbox_inches='tight')
plt.show()
```

```python
currcount = gdp['Currency'].value_counts().reset_index()
currmean = gdp.groupby('Currency')['GDP'].mean().reset_index()
curr = currcount.merge(currmean, how = 'inner', left_on = 'index', right_on = 'Currency').iloc[:,[0,1,3]]
curr.columns = ['Currency', 'Count', 'GDP']
fig = px.scatter(curr, x = 'Count', y = 'GDP', hover_data = 'Currency',
                 title = 'Average GDP per Capita Grouped by Primary Currency',
                 labels = {'Count': 'Number of Countries', 'GDP': 'Average Real GDP per Capita'})
fig.write_html('currency.html')
```

```python
sns.boxplot(gdp, x = 'Region', y = 'GDP')
plt.xlabel('United Nations Region')
plt.ylabel('Real GDP per Capita (PPP) in US Dollars')
plt.title('GDP per Capita Spread by UN Region')
plt.savefig('boxplot.png', bbox_inches = 'tight')
plt.show()
```

```python
sns.lmplot(data = gdp, x = 'Density', y = 'GDP', scatter_kws = {'s': 10}, ci =None) #plot ratio to gdp scatterplot with fits for each Region
plt.title('Economic Prosperity Relative to Population Density')
plt.xlabel('Population per Square Mile')
plt.ylabel('Real GDP per Capita (PPP) in US Dollars')
plt.savefig('density.png', bbox_inches='tight')
plt.show()
```

```python
sns.lmplot(data = gdp, x = 'Density', y = 'GDP', hue = 'Region', scatter_kws = {'s': 10}, ci =None) #plot ratio to gdp scatterplot with fits for each Region
plt.title('Economic Prosperity Relative to Population Density by Region')
plt.xlabel('Population per Square Mile')
plt.ylabel('Real GDP per Capita (PPP) in US Dollars')
plt.xlim(0, 6000)
plt.savefig('densityregion.png', bbox_inches='tight')
plt.show()
```

```python
sns.heatmap(gdp.corr(), annot = True)
plt.title('Correlation Heatmap')
plt.savefig('correlation.png', bbox_inches='tight')
plt.show()
```

```python
sns.histplot(x = 'GDP', data = gdp)
plt.title('Worldwide GDP Spread Histogram')
plt.xlabel('Real GDP per Capita (PPP) in US Dollars')
plt.ylabel('Number of Countries')
plt.savefig('gdphist.pngb', bbox_inches='tight')
plt.show()
```
